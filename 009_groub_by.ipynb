# Group by
V tomto videu sme sa rozprávali o group by funkcii založenej na princípe split-apply-combine

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv("dataset/train.csv")
df.shape
#as_index=False, sort=True
```

```python
df.MSZoning.value_counts()
```

```python
# Group by funkcia nevracia priamo zoskupenie ale generator
df.groupby("MSZoning")
```

```python
# Nasledne na group  by objekt vieme aplikovat bud build in funkciu alebo funkciu apply
df.groupby("MSZoning").mean()
```

```python
df.groupby("MSZoning").count()
```

```python
# Group by vieme aplikovat aj na viacero stlpcov sucasne, presne ako v pripade group by v SQL 
df.Alley.value_counts()
```

```python
df.groupby(["MSZoning", "Alley"]).count()
```

```python
# Pre zachovanie stpcov z group by funkcie pouzijeme parameter as_index rovny False
df.groupby(["MSZoning", "Alley"], as_index=False).count()
```

```python
# Vysledok group by vieme aj zoradit podla group by stlpcov s parametrom sort = True
df.groupby(["MSZoning", "Alley"], sort=True).count()
```

```python
# Na group by objekt vieme aplikovat aj vlastnu funkciu s pouzitim funkcie vyssieho radu apply
# platia tu pravidla z predchadzajucej lekcie 

def total_area(group_df):
    """
    Ked chcem z vlastnej funkcie vratit viac stlpcov staci vratit pandas Series objekt kde index bude tvorit nazvy stlpcov
    """
    return pd.Series([(group_df.LotArea + group_df.LotFrontage).sum(), (group_df.LotArea + group_df.LotFrontage).mean()], index=["sum", "mean"])

df.groupby(["MSZoning"]).apply(total_area)
```

```python
# Pouzite group by apply spolu s lambda funkciou
df.groupby(["MSZoning"]).apply(lambda group_df: pd.Series([
    (group_df.LotArea + group_df.LotFrontage).sum(), 
    (group_df.LotArea + group_df.LotFrontage).mean()], index=["sum", "mean"]))
```
