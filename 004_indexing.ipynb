# Pandas copy vs view
- ked pracujem s loc a iloc tak pracujem s __view__ alebo s referenciou do povodneho dataframu ked zmenim nieco cez iloc menim aj povodny df.
- toto neplati vzdy pri pristupe cez hranate zatvorky. Tieto v niektorych pripadoch vracaju __copy__ co su vlastne skopirovane data z vyberu a teda su ako keby novy dataframe.

```python
# Nulovanie cez hranate zatvorky a ich retazenie sposobuje warning !!! 
# Warning hovory ze niekedy sa sprava pristup cez zatvorky ako view a niekedy nie a treba si dat na to pozor
df[df.SalePrice > 181500].SalePrice = 0 # toto je zly pristup 
```

```python
# Hodnoty sa vsak nezmenili lebo sme pracovali s kopiou nie s view. TOTO JE ZLE PRETO RADSEJ POUZIVAME loc a iloc!!!
df.head()
```

```python
# Nepomoze ani ked to oddelime na viac prikazov
temp_df = df[df.SalePrice > 181500] # lebo toto je uz kopia a nie je to view.
temp_df.SalePrice = 0
df.head()
```

```python
# MAGIC, ked zmenime poradie VoilÃ , zrazu to funguje. LOOOL 
df.SalePrice[df.SalePrice > 181500] = 0
df.head()
```

```python
# Znovu nacitat data
df = pd.read_csv("dataset/train.csv")

# Riesenie vsetkeho je pouzivat loc iloc lebo vzdy vracia view
df.loc[df.SalePrice > 181500, "SalePrice"] = 10
df
```

```python
# Chainovanie loc a iloc nie je dobre. Efektivnejsie je mat co najmenej loc iloc
df.loc[df.SalePrice > 181500, :].loc[df.MSSubClass > 30, :] # toto je zle 
df.loc[(df.SalePrice > 181500) & (df.MSSubClass > 30), :] # toto je cool
```
