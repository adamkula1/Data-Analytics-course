```python
def total_area(group_df):
    """
    Ked chcem z vlastnej funkcie vratit viac stlpcov staci vratit pandas Series objekt kde index bude tvorit nazvy stlpcov
    """
    return pd.Series([(group_df.LotArea + group_df.LotFrontage).sum(), (group_df.LotArea + group_df.LotFrontage).mean()], index=["sum", "mean"])

df.groupby(["MSZoning"]).apply(total_area)
```
