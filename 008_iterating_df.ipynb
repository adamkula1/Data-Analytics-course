# Prechadzanie dataframom
V tomto videu sme si ukázali ako vieme prechádzať cez dataframe s vyžitím dvoch prístupov
- Iteratívny prístup pomocou funkcie iterrows.
- Funkcionálny prístup s použitím funkice vyššieho rádu apply.

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv("dataset/train.csv")
df.head()
```

```python
# Funkcia iterrows vracia generátor
df.iterrows()
```

```python
# Ak chceme vidiet obsah pouzijeme funkciu list. [:2] vrati len prve dva prvky aby nebol prilis dlhy vypis
list(df.iterrows())[:2]
```

```python
# Prvky iterrows su tuple tvorene indexom a hodnotou riadku z dataframu
type(list(df.iterrows())[0])
```

```python
# Pomocou iterrows vieme prechadzat cez dataframe
for index, row in list(df.iterrows())[:2]:
    print(f"Index is {index} mand values are {row}")
```

```python
# Pomocou iterrows vieme menit dataframe
for index, row in df.iterrows():
    df.loc[index, "Id"] = 0
df.head()
```

# Prechadzanie pomocou funkcie apply

```python
df = pd.read_csv("dataset/train.csv")
df.head()
```

```python
def print_row(row):
    print(f"Index is {row.name} mand values are {row}")
    
def edit_id(row):
    row.Id = 0
    return row

df.loc[:, ["Id", "SalePrice"]].apply(edit_id, axis=1)
```

```python
# Pomocou apply funkcie vieme pouzit aj takzvanu lambda funkciu
df.loc[:2, ["Id", "SalePrice"]].apply(lambda row: print(f"Index is {row.name} mand values are {row}"), axis=1)
```

```python
# Apply funkcia nemeniu povodny dataframe pokial nenastavime parameter inplace=True (defaultne je na False)
df
```

```python
# Pomocou apply funkcie vieme pouzit aj funkciu s parametrami ktore posleme pomocou tuple
def process_prices(row, substractror, x,y,z):
    row.SalePrice = row.SalePrice - substractror
    return row

df.loc[:, ["Id", "SalePrice"]].apply(process_prices, args=(20,1,2,3), axis=1)
```
